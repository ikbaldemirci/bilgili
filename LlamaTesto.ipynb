# **Google Drive Bağlantısı**

```python
from google.colab import drive

# 1. Google Drive'ı bağlama
drive.mount('/content/drive')

# 2. Txt dosyasını okuma
file_path = '/content/drive/My Drive/bilgili/bilgili.txt'  # Dosyanızı Google Drive'a taşıdıktan sonra bu yolu kontrol edin.
with open(file_path, 'r', encoding='utf-8') as file:
    data = file.read()
```

```python
import re

# Veriyi işleyin
info_dict = {}
current_section = None
for line in data.split('\n'):
    if line.strip():
        if re.match(r"[A-Z\s]+", line):  # Büyük harflerle yazılmış bölümler
            current_section = line.strip()
            info_dict[current_section] = []
        else:
            if current_section:
                info_dict[current_section].append(line.strip())

print("Veriler başarıyla işlendi.")
```

```python
# Soru-Cevap Fonksiyonu
def get_info(query):
    for section, details in info_dict.items():
        if query.lower() in section.lower():
            return f"{section}:\n" + "\n".join(details)
    return "Maalesef, bu konu hakkında bilgi bulunamadı."

# Kullanıcıdan girdi alma ve cevaplama
while True:
    user_query = input("Hangi bilgiye ulaşmak istiyorsunuz? (Çıkmak için 'exit' yazın): ")
    if user_query.lower() == 'exit':
        break
    response = get_info(user_query)
    print(response)
```

# .txt dosyasını yapılandırma ve kaydetme adımları

```python
import re
import json

# Dosya yolunu belirtin
file_path = '/content/drive/My Drive/bilgili/bilgili.txt'

# Veriyi işlemek için sözlük yapısı
structured_data = {}
current_section = None
current_subsection = None

# Dosyayı oku ve işlem yap
with open(file_path, 'r', encoding='utf-8') as file:
    for line in file:
        line = line.strip()
        if not line:
            continue  # Boş satırları atla

        # Ana başlıklar (fakülte veya bölüm isimleri)
        if re.match(r"^[A-ZÇŞĞÜÖİ\s]+$", line):
            current_section = line.strip()
            structured_data[current_section] = {}
            current_subsection = None
        # Alt başlıklar (örneğin, fakülte içindeki bölümler)
        elif "(" in line and ")" in line:
            if current_section:  # Ana başlık varsa işlem yap
                current_subsection = line.strip()
                structured_data[current_section][current_subsection] = {}
        # Yıl ve ücret bilgisi
        elif re.match(r"^\d{4}.*", line):
            if current_section:  # Ana başlık olmadan işlem yapma
                year_info, fee = line.rsplit(" ", 1)
                if current_subsection:  # Alt başlık varsa ona ekle
                    structured_data[current_section][current_subsection][year_info] = fee
                else:  # Alt başlık yoksa doğrudan ana başlığa ekle
                    structured_data[current_section][year_info] = fee

# Veriyi kontrol et
print(json.dumps(structured_data, indent=4, ensure_ascii=False))
```

```python
output_file_path = '/content/drive/My Drive/bilgili/bilgili_yapilandirilmis.json'
with open(output_file_path, 'w', encoding='utf-8') as json_file:
    json.dump(structured_data, json_file, indent=4, ensure_ascii=False)

print(f"Yapılandırılmış dosya kaydedildi: {output_file_path}")
```

```python
with open(output_file_path, 'r', encoding='utf-8') as json_file:
    structured_data = json.load(json_file)

#kontrol
print(json.dumps(structured_data, indent=4, ensure_ascii=False))
```
